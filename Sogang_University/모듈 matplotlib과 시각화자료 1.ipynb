# 기술통계(descriptive statistics)

- 표본의 분포적 특성 요약, 즉, 개별 표본 / 데이터 자체에 관심
- 양적 척도
1. 정규분포 : 중심경향도 평균, 산포도 표준편차
2. 비정규분포 : 중심경향도 중앙값, 산포도 사분위수 범위(IQR:Interquartile Range)
3. 시각자료 : 상자그림(box plot), 산점도(scatter plot) 등
- 질적 척도
1. 빈도 및 비율
2. 파이차트(pie chart), 막대 그래프(bar plot) 등

## 데이터 분석에서 시각 기법의 효과
- 시각 정보
1. 압축된 원리/법칙에 대한 이해 또는 복잡한 양적 수치 자료에 대한 이해를 쉽게 만든다.
2. 정량적 정보를 정성적 정보로 변환
3. 수치에서 제시되는 것 이상의 정보를 제공하여 통합적 통찰을 통해 추세/패턴 탐색 용이
4. 데이터 사이언스에서는 시각 분석을 통한 패턴 탐색 중시

## plot 출력 : inline

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot([1,2,3],[5,2,4])
```

```python
plt.plot([5,2,4]) # 0, 1, 2에 맞춰서 값
```

## plot: ndarray

```python
import numpy as np
plt.plot(np.arange(0,5,2) * np.arange(3))
```

## plot: Series

```python
import pandas as pd
plt.plot(pd.Series(np.arange(0, 5, 2) * np.arange(3),
                  index = range(1,4)))
```

```python
a = pd.Series(np.arange(0,5,2) * np.arange(3), index = range(1,4))
a.index.name = 'x'
a.plot()
```

```python
plt.plot(range(6))
pd.Series(np.arange(6,0,-1), index= range(1,7)).plot()
```

## plot: Data Frame

```python
plt.plot(pd.DataFrame({'A': dict([(1,3),(3,4),(5,1)]),
                      'B': dict([(1,1),(3,2),(5,5)])}))
```

```python
a = pd.DataFrame({'A': dict([(1,3),(3,4),(5,1)]),
                      'B': dict([(1,1),(3,2),(5,5)])})
a.index.name = 'X'
a.columns.name = 'CLASS'
a.plot()
```

## plot 함수와 데이터 구조

### 차트 만들기 : 기본

```python
plt.plot(range(5), range(6,1,-1),'.b')
```

1. Markers
- Character description<br>
1) '.'       point marker<br>
2) ','       pixel marker<br>
3) 'o'       circle marker<br>
<br>
2. Line Styles
- Character description<br>
1) '-'       solid line style<br>
2) '--'      dashed line style<br>
3) '-.'      dash-dot line style<br>
4) ':'       dotted line style<br>
3. color
- Character color<br>
1) 'b'       blue<br>
2) 'g'       green<br>
3) 'r'       red<br>
4) 'c'       cyan<br>
5) 'm'       magenta<br>
6) 'y'       yellow<br>
7) 'k'       black<br>
8) 'w'       white<br>

```python
plt.plot(range(5), range(6,1,-1),
        color='r', linestyle='-.', marker='o')
```

```python
plt.plot(range(5), range(6,1,-1), 'k-.',
        range(5), range(6,1,-1), 'ro',
        range(5), range(7,2,-1), 'g--',
        range(5), range(7,2,-1), 'bo')
```

# 연습문제 1

```python
import numpy as np
import matplotlib.pyplot as plt
plt.plot(np.arange(0, 4, 0.5), list(range(4)) * 2, 'ko-',
        np.arange(0, 4, 0.5), list(range(4,0,-1)) * 2, 'ko--')
```

# 연습문제 2

```python
import numpy as np
import matplotlib.pyplot as plt
a = pd.DataFrame({'A': list(range(4)) * 2,
                 'B': list(range(4,0,-1)) * 2},
                index = np.arange(0,4,0.5))
a.index.name = 'size'
a.columns.name = 'time'
a.plot(linestyle='--', marker='o')
```
