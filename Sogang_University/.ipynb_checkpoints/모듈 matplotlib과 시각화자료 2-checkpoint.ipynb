# axis

```python
import matplotlib.pyplot as plt 
plt.plot(range(5),range(6,1,-1),'bo')
plt.axis([-2,6,1,8])  #출력되는 x축 y축 구간 정의
```

## xlabel, ylabel

```python
plt.plot(range(5), range(6,1,-1), ':r.')
# 한글 깨짐 해결
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.xlabel('X-축')
plt.ylabel('Y-축')
```

## xticks, yticks

```python
from numpy.random import randn
data = randn(1000).cumsum()
# 마이너스 기호
plt.rcParams['axes.unicode_minus'] = False
plt.plot(data,'-r')
plt.xticks([0,250,500,750,1000])
plt.yticks(range(-20,21,10))
```

```python
plt.plot(data,'-r')
plt.xticks([0,250,500,750,1000], list('ABCDE'),
          rotation = 45, fontsize = 'medium', style='italic')
```

## title, legend

```python
plt.plot(randn(100).cumsum(), 'r-')
plt.plot(randn(100).cumsum(), 'b:')
plt.title('두 추세선')
plt.legend(['1차','2차'], loc = 'best')
```

## 연습문제 1

```python
from numpy.random import randn
A = randn(100).cumsum()
B = randn(100).cumsum()
plt.plot(A,'r-')
plt.plot(B,'b:')
plt.title('A vs B')
plt.legend(['A','B'], loc = 'best')
plt.xticks(range(0,101,10), range(1900,2001,10),
          rotation=90)
plt.xlabel('YEAR')
```
