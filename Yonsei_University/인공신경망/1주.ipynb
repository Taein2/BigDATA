```python
# 트랜잭션의 총 갯수
tr_num = 50

# 한 트랜잭션 내의 항목의 갯수 (범위)
tr_len = list(range(2, 6)) # 2 ~ 5

# 항목이 가질 수 있는 값의 범위
item_domain = list(range(1, 21)) # 1 ~ 20

print(tr_len)
print(item_domain)
```

```python
data = []

for a in range(tr_num):
    # choice는 인자가 하나이기에 꺼낼때마다 갯수가 달라질거임
    c = random.sample(item_domain,random.choice(tr_len)) 
    c.sort()
    data.append(c)
```

```python
print(len(data))
```

```python
import apyori as ap

rules = list(ap.apriori(data, min_support = 0.1, min_confidence = 0.2))
```

```python
for r in rules:
    if len(r[0]) > 1:
        support = r[1]
        ar = list(r[2])
        for a in ar:
            print("{} ==> {}".format(a[0],a[1]))
            print("support = {}".format(r[1]))
            print("confidence = {}".format(a[2]))
            print("list = {}", a[3])
            print()
```

```python
# 데이터 프레임 
import pandas as pd

result = []

for r in rules:
    if len(r[0]) > 1:
        support = r[1]
        ar = list(r[2])
        for a in ar:
            row = [set(a[0]), "==>", set(a[1]), support , a[2], a[3]]
            result.append(row)
col = ["선행절","==>","후행절","지지도","신뢰도","향상도"]
df = pd.DataFrame(result, columns = col, index = range(1,len(result)+1))
df
```
