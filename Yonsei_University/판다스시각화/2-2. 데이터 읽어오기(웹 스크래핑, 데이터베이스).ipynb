### 3) Selenium 라이브러리 활용  

* selenium 설치: conda install selenium
* chromedriver 다운로드 : https://sites.google.com/a/chromium.org/chromedriver/downloads

```python
# selenium으로 위의 에러 문제를 해결해보자!
from selenium import webdriver
driver = webdriver.Chrome('chromedriver')
```

```python
driver.implicitly_wait(3)
```

```python
삼성전자url = 'https://finance.naver.com/item/sise_day.nhn?code=005930&page='
```

```python
driver.get(삼성전자url)
```

```python
html = driver.page_source
```

```python
data = pd.read_html(html,encoding='cp949')
```

```python
삼성_일별 = data[0][data[0].날짜.notnull()]
```

```python
삼성_일별
```

```python
#실습 모든 기간의 삼성전자 일별 데이터를 스크래핑해서 하나의 데이터 프레임 생성
일별주가 = []
삼성전자url = 'https://finance.naver.com/item/sise_day.nhn?code=005930&page='
for i in range(1,3): # 편의상 1,2만
    full_url = 삼성전자url + str(i)
    driver.get(full_url)
    html = driver.page_source
    data = pd.read_html(html,encoding='cp949')
    삼성_일별 = data[0][data[0].날짜.notnull()]
    일별주가.append(삼성_일별)
```

```python
일별주가[1]
```

```python
삼성전자_일별_최종데이터 = pd.concat(일별주가)
```

```python
삼성전자_일별_최종데이터.to_excel('data/삼성전자_일별주가.xlsx', encoding='cp949')
```

## 4. 데이터베이스에서 읽기
### 파이썬에 내장(embedded)된 sqlite 사용

```python
import sqlite3
```

```python
#테이블 생성 SQL 정의
테이블생성 = 'CREATE TABLE locations \
                (a varchar[20], b varchar[20], c integer)'
```

```python
#sqlite 접속
conn = sqlite3.connect('temp.db')
```

```python
#테이블 생성 SQL 수행 
conn.execute(테이블생성)
```

```python
# commit() 수행. - SQL 수행 후, commit을 해야만, 데이터베이스에 SQL 구문이 적용됨 
conn.commit()
```

```python
#생성된 테이블에 저장할 데이터 정의
data = [['서울','강남', 15]
       ,['서울','강동', 10]
       ,['서울','동대문', 7]
       ,['서울','양천', 12]
       ,['성남','분당', 12]
       ,['수원','영통', 15]
       ,['성남','중원', 11]
       ,['수원','팔달', 14]
       ]
```

```python
# 데이터 적재 SQL 정의 
삽입 = 'insert into locations values (?,?,?)'
```

```python
#데이터 적재 SQL 수행 
conn.executemany(삽입,data)
```

```python
# commit() 수행 
conn.commit()
```

```python
#Select SQL 실행1
pd.read_sql('select * from locations',conn)
```

```python
#Select SQL 실행2
pd.read_sql('select a,sum(c) from locations group by a',conn)
```

```python
# 데이터베이스의 테이블 크기가 큰 경우에는 cursor 사용하기
cursor = conn.execute('select * from locations')
cursor
```
