[실습] 서울시 유동인구 데이터에서 조사지점 정보들을 MarkerCluster로 추가하기
 - data/서울시유동인구/4_유동인구_조사지점정보_2015.xlsx 활용
 - 조사구분이 '본조사'인 데이터만 대상으로 함

```python
import pandas as pd
```

```python
# 데이터 읽기
측정위치 = pd.read_excel('data/서울시유동인구/4_유동인구_조사지점정보_2015.xlsx', 
                      skiprows = [0,1,3])
측정위치.head()
```

```python
측정위치 = 측정위치[['조사지점명','X좌표','Y좌표']]
```

```python
측정위치.X좌표.notnull()
```

```python
측정위치.head(2)
```

```python
# 위치 좌표 변환
# !pip install pyproj
```

```python
# 좌표계 변환
from pyproj import Proj, transform
```

```python
def convert_loc(x,y):
    inProj = Proj(init = 'epsg:5181')
    outProj = Proj(init = 'epsg:4326')
    x2, y2 = transform(inProj, outProj,x,y)
    return [y2,x2]
```

```python
convert_loc(96423.97707,455511.52968)
```

```python
측정위치['위경도'] = 측정위치.apply(lambda 지점: convert_loc(지점.X좌표,지점.Y좌표), axis=1)
```

```python
측정위치.head()
```

```python
map3 = folium.Map([37.566535, 126.97796919999996],zoom_start=13,
                 zoom_control = False, control_scal=True,
                 tiles = 'https://tiles.stadiamaps.com/tiles/outdoors/{z}/{x}/{y}{r}.png',
                 attr = '&copy; <a href="https://stadiamaps.com/">Stadia Maps</a>, &copy; <a href="https://openmaptiles.org/">OpenMapTiles</a> &copy; <a href="http://openstreetmap.org">OpenStreetMap</a> contributors')
map3
```

```python
marker_cluster = MarkerCluster().add_to(map3)
for idx,val in 측정위치.iterrows():
    folium.Marker(location = val.위경도, popup = folium.Popup(val.조사지점명, max_width = 300))\
    .add_to(marker_cluster)
```
